```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.stats import shapiro
import scipy.stats
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import bartlett
from statsmodels.api import Logit
```

```python
df = pd.read_csv("최종데이터.csv",encoding = 'euc - kr')
df
```

```python
df.info()
```

```python
생산일자_pivot_df= df.pivot_table(index='생산일자', columns='출하완료여부', fill_value=0, aggfunc='size')
생산일자_pivot_df
```

```python
chi,pval,dof,expected = stats.chi2_contingency(생산일자_pivot_df)

print("chi-square test")
print("chisq : %.3f"%chi)
print("p : %f"%pval)
print("dgree of freedom : %.3f"%dof)
print("expected value : \n{}".format(expected.round(3))) # p-value < 0.5 대립가설 채택
```
